# Assignment: Trace all the numeric variables from the Dataset and make them numeric

```python
# # numeric Dataset
# - Size
# - Install
# - Price
```

```python
df['Installs'] = df['Installs'].str.replace('+', '')
df['Installs'] = df['Installs'].str.replace(',', '')
df['Price'] = df['Price'].str.replace('$', '')
df
```

```python
df['Size'].replace('Varies with device', np.nan, inplace=True)
```

```python
# Function to convert Size to MB
def convert_to_mb(size_str):
    if isinstance(size_str, str):
        if 'k' in size_str:
            return float(size_str.replace('k', '')) / 1024
        elif 'M' in size_str:
            return float(size_str.replace('M', ''))
        else:
            return None  # Handle cases where format is unknown
    else:
        return size_str  # Return the value as-is if already numeric

# Apply the conversion function
df['Size'] = df['Size'].apply(convert_to_mb)

# Now 'Size' column is in MB, numeric format
print(df)
```

```python
df['Size'].fillna(df['Size'].mean(), inplace=True)
```

```python
# Convert type of Installs and Price from object to float

df['Installs'] = df['Installs'].astype(float)
df['Price'] = df['Price'].astype(float)
```
